### üìê Exercise 3: Correlation and Regression Analysis
Perform statistical analysis on numeric relationships.


```python
# Exercise 3: Correlation and Regression Analysis
# Write the solution to Exercise 3 here

# Your task:
# A) Compute the correlation matrix among numeric variables (Speed, Length, and Width)
#
# B) Perform a simple linear regression analysis to examine the relationship between:
#    - Independent variable (X): Vessel Length
#    - Dependent variable (Y): Speed
#
# C) Plot the regression line and interpret the relationship
```

```python
columns = ["Speed", "Length","Width"]

matrix = shipsdf[columns].corr()
matrix
```

```python
import numpy as np
```

```python
X = shipsdf["Length"]
Y = shipsdf["Speed"]
mask = ~X.isna() & ~Y.isna() 
X_clean = X[mask]
Y_clean = Y[mask]

slope, intercept = np.polyfit(X_clean, Y_clean, 1)
print(f"Regression equation: Speed = {intercept:.2f} + {slope:.2f} * Length")
```

```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import matplotlib.colors as colors
```

```python
plt.scatter(X_clean, Y_clean, label='Data Points', alpha=0.6)
plt.plot(X_clean, slope * X_clean + intercept, color='red', label='Regression Line')
plt.xlabel('Length')
plt.ylabel('Speed')
plt.title('Relationship between Speed and Length')
plt.legend(loc=(1.02, 0))
plt.show()
```

### üîç Exercise 6: Analytical Queries - Part A
Query and analyze employee data by department and location.

```python
# Exercise 6A: Analytical Queries - Departmental & Geographic Analysis
# Write the solution to Exercise 6A here

# Your task:
# Perform the following analytical queries and display results as Pandas DataFrames.
# Write each query BOTH in raw SQL AND show the DataFrame output.
#
# Query A) Average salary per department
#
# Query B) Number of employees per country
#
# Query C) Average bonus percentage by business unit
#
# Query D) List of employees hired after 2020, sorted by salary (descending)
#
# Query E) The top 5 highest-paid employees overall
```

```python
#-------------------------A---------------------
```

```python
# SQL
sql_a = """
SELECT Department, AVG(AnnualSalary) as avg_salary
FROM EMPLOYEE
GROUP BY Department
ORDER BY avg_salary DESC
"""
print("SQL a:\n", pd.read_sql(sql_a, sqlite3.connect("employees.db")))

# pandas
pandas_a = (employees
            .groupby("Department")["AnnualSalary"]
            .mean()
            .reset_index()
            .sort_values("AnnualSalary", ascending=False)
           )
print("pandas a:\n", pandas_a)
```

```python
#-------------------------B---------------------
```

```python
# SQL
sql_b = """
SELECT Country, COUNT(*) as employee_count 
FROM EMPLOYEE 
GROUP BY Country 
ORDER BY employee_count DESC
"""
print("SQL b:")
print(pd.read_sql(sql_b, sqlite3.connect("employees.db")))

# pandas
print("\n pandas b:")
pandas_b = employees["Country"].value_counts().reset_index()
pandas_b.columns = ["Country", "employee_count"]
print(pandas_b.sort_values("employee_count", ascending=False))
```

```python
#-------------------------C---------------------
```

```python
# SQL 
sql_c = """
SELECT BusinessUnit,
       ROUND(AVG(BonusPercent)*100, 1) || '%' as avg_bonus_pct
FROM EMPLOYEE
GROUP BY BusinessUnit
ORDER BY AVG(BonusPercent) DESC
"""
print("SQL c (%):")
print(pd.read_sql(sql_c, sqlite3.connect("employees.db")))

# pandas 
print("\npandas c (%):")
pandas_c = (employees
    .groupby("BusinessUnit")["BonusPercent"]       
    .mean()                                         
    .mul(100)                                      
    .round(1)                                       
    .reset_index()                                  
    .assign(BonusPercent=lambda x: x["BonusPercent"] 
            .astype(str) + '%')                     
    .sort_values("BonusPercent",                    
                 key=lambda x: x.str.rstrip('%').astype(float),  
                 ascending=False))                      

print(pandas_c)
```

```python
#-------------------------D---------------------
```

```python
# SQL
sql_d = """
SELECT FullName, JobTitle, AnnualSalary, HireDate
FROM EMPLOYEE
WHERE HireDate > '2020-12-31'
ORDER BY AnnualSalary DESC
"""
print("SQL d:")
print(pd.read_sql(sql_d, sqlite3.connect("employees.db")))

# pandas
print("\npandas d:")
pandas_d = (employees[employees["HireDate"] > "2020-12-31"][["FullName", "JobTitle", "AnnualSalary", "HireDate"]]
.sort_values("AnnualSalary", ascending=False))

print(pandas_d)
```

```python
#-------------------------E---------------------
```

```python
# SQL
sql_e = "SELECT FullName, JobTitle, AnnualSalary FROM EMPLOYEE ORDER BY AnnualSalary DESC LIMIT 5"
print("SQL e:")
print(pd.read_sql(sql_e, sqlite3.connect("employees.db")))

# pandas
print("\npandas e:")
pandas_e = employees.nlargest(5, "AnnualSalary")[["FullName", "JobTitle", "AnnualSalary"]]
print(pandas_e)
```

### üìä Exercise 6B: Analytical Queries - Part B
Query and analyze employee demographics and HR metrics.


```python
# Exercise 6B: Analytical Queries - Demographics & HR Metrics
# Write the solution to Exercise 6B here

# Your task:
# Perform the following analytical queries and display results as Pandas DataFrames.
# Write each query BOTH in raw SQL AND show the DataFrame output.
#
# Query F) The average age of employees per department
#
# Query G) Number of active vs. exited employees
#          (based on ExitDate being NULL or not)
#
# Query H) Average annual salary per gender and country
#
# Query I) Number of employees in each ethnicity category
#
# Query J) Department(s) with the highest average bonus percentage
```

```python
#-------------------------F---------------------
```

```python
# SQL
sql_f = "SELECT Department, AVG(Age) as avg_age FROM EMPLOYEE GROUP BY Department ORDER BY avg_age DESC"
print("SQL f:")
print(pd.read_sql(sql_f, sqlite3.connect("employees.db")))

# pandas
print("\npandas f:")
pandas_f = employees.groupby("Department")["Age"].mean().reset_index().sort_values("Age", ascending=False).round(1)
print(pandas_f)
```

```python
#-------------------------G---------------------
```

```python
# SQL
sql_g = """
SELECT
    CASE WHEN ExitDate IS NULL OR ExitDate = '' THEN 'Active' ELSE 'Exited' END as status,
    COUNT(*) as count
FROM EMPLOYEE
GROUP BY CASE WHEN ExitDate IS NULL OR ExitDate = '' THEN 'Active' ELSE 'Exited' END
"""
print("SQL g:")
print(pd.read_sql(sql_g, sqlite3.connect("employees.db")))

# pandas
print("\npandas g:")
pandas_g = employees["ExitDate"].apply(lambda x: "Active" if pd.isna(x) or x == "" else "Exited").value_counts().reset_index()
pandas_g.columns = ["status", "count"]
print(pandas_g)
```

```python
#-------------------------H---------------------
```

```python
# SQL
sql_h = """
SELECT Gender, Country, AVG(AnnualSalary) as avg_salary
FROM EMPLOYEE
GROUP BY Gender, Country
ORDER BY avg_salary DESC
"""
print("SQL h:")
print(pd.read_sql(sql_h, sqlite3.connect("employees.db")))

# pandas
print("\npandas h:")
pandas_h = (employees.groupby(["Gender", "Country"])["AnnualSalary"]
                    .mean()
                    .reset_index()
                    .sort_values("AnnualSalary", ascending=False))
print(pandas_h)
```

```python
#-------------------------I---------------------
```

```python
# SQL
sql_i = "SELECT Ethnicity, COUNT(*) as count FROM EMPLOYEE GROUP BY Ethnicity ORDER BY count DESC"
print("SQL i:")
print(pd.read_sql(sql_i, sqlite3.connect("employees.db")))

# pandas
print("\npandas i:")
pandas_i = employees["Ethnicity"].value_counts().reset_index()
pandas_i.columns = ["Ethnicity", "count"]
print(pandas_i)
```

```python
#-------------------------J---------------------
```

```python
# SQL 
sql_j = """
SELECT Department,
       ROUND(AVG(BonusPercent)*100, 1) || '%' as avg_bonus_pct
FROM EMPLOYEE
GROUP BY Department
ORDER BY AVG(BonusPercent) DESC
LIMIT 1
"""
print("SQL j (%):")
print(pd.read_sql(sql_j, sqlite3.connect("As_employees.db")))

# pandas 
print("\npandas j (%):")
pandas_j = (employees.groupby("Department")["BonusPercent"]
            .mean()
            .mul(100)
            .round(1)
            .reset_index()
            .assign(BonusPercent=lambda x: x["BonusPercent"].astype(str) + '%')
            .sort_values("BonusPercent", key=lambda x: x.str.rstrip('%').astype(float), ascending=False)
            .head(1))
print(pandas_j)
```
